```python
import matplotlib.pyplot as plt

sprints = ['Sprint 1', 'Sprint 2', 'Sprint 3', 'Sprint 4', 'Sprint 5', 'Sprint 6']
envio_exito = [70, 80, 85, 87, 90, 95]
errores_validacion = [10, 8, 7, 5, 3, 2]
verificacion_layout = [60, 70, 75, 80, 85, 90]
verificacion_funcionalidad = [50, 65, 75, 80, 85, 90]
verificacion_compatibilidad = [55, 65, 75, 80, 85, 90]
documentacion = [50, 60, 70, 75, 80, 85]
satisfaccion_usuario = [10, 15, 20, 25, 30, 35]

fig, axs = plt.subplots(2, 3, figsize=(15, 10))


axs[0, 0].plot(sprints, envio_exito, marker='o', color='blue')
axs[0, 0].set_title('Tasa de xito de envo')
axs[0, 0].set_ylabel('Porcentaje')

axs[0, 1].plot(sprints, errores_validacion, marker='o', color='red')
axs[0, 1].set_title('Nmero de errores de validacin')
axs[0, 1].set_ylabel('Errores')


axs[0, 2].plot(sprints, verificacion_layout, marker='o', color='green')
axs[0, 2].set_title('Verificacin de layout y diseo')
axs[0, 2].set_ylabel('Porcentaje')


axs[1, 0].plot(sprints, verificacion_funcionalidad, marker='o', color='cyan')
axs[1, 0].set_title('Verificacin de funcionalidad')
axs[1, 0].set_ylabel('Porcentaje')

axs[1, 1].plot(sprints, verificacion_compatibilidad, marker='o', color='magenta')
axs[1, 1].set_title('Verificacin de compatibilidad')
axs[1, 1].set_ylabel('Porcentaje')

axs[1, 2].plot(sprints, documentacion, marker='o', color='yellow', label='Documentacin')
axs[1, 2].plot(sprints, satisfaccion_usuario, marker='o', color='black', label='Satisfaccin del usuario')
axs[1, 2].set_title('Documentacin y satisfaccin del usuario')
axs[1, 2].set_ylabel('Puntuacin / Porcentaje')
axs[1, 2].legend()

for ax in axs.flat:
    ax.set_xlabel('Sprints')

plt.tight_layout()
plt.show()
```
